## 3. Data Description

This analysis uses three datasets containing information about business hypotheses, user visits, and customer orders during the A/B test.

### hypotheses_us.csv

This dataset contains the list of marketing and product hypotheses designed to increase the revenue of the online store. Each hypothesis was evaluated using four criteria that support prioritization frameworks.

Variables:

- **Hypothesis** — brief description of the proposed improvement or experiment.
- **Reach** — estimated number of users affected by the change (scale from 1 to 10).
- **Impact** — expected impact on user behavior if the hypothesis is implemented (scale from 1 to 10).
- **Confidence** — level of confidence in the hypothesis (scale from 1 to 10).
- **Effort** — estimated resources required to implement and test the hypothesis (scale from 1 to 10).

---

### orders_us.csv

This dataset contains information about orders placed during the A/B test.

Variables:

- **transactionId** — unique identifier of each order.
- **visitorId** — unique identifier of the user who placed the order.
- **date** — date when the order was made.
- **revenue** — revenue generated by the order.
- **group** — A/B test group to which the user belongs (A or B).

---

### visits_us.csv

This dataset contains the number of daily visits for each test group.

Variables:

- **date** — date of the recorded visits.
- **group** — A/B test group (A or B).
- **visits** — total number of visits recorded on that date for the corresponding group.

---

Before performing the analysis, the datasets will be inspected and preprocessed to ensure data quality, including checking for missing values, incorrect data types, duplicates, and potential inconsistencies such as users appearing in both experimental groups.

## 5. Hypothesis Prioritization

### 5.1 ICE Framework

```python
hypotheses.columns = hypotheses.columns.str.lower()
hypotheses.head()
```

```python
hypotheses['ICE']=hypotheses['impact']*hypotheses['confidence']/hypotheses['effort']
hypotheses.sort_values(by='ICE', ascending=False).round(2)
```

### 5.2 RICE Framework

```python
hypotheses['RICE']=hypotheses['reach']*hypotheses['impact']*hypotheses['confidence']/hypotheses['effort']

hypotheses.sort_values(by='RICE', ascending=False).round(2)
```

### 5.3 ICE vs RICE Comparison

### ICE

After calculating the ICE scores, the hypotheses were ranked in descending order of priority.  
The five most promising hypotheses were: **8, 0, 7, 6, and 2**.

These hypotheses combine **high impact and confidence with relatively low implementation effort**, making them strong candidates for initial testing.

### RICE

After calculating the RICE scores, the hypotheses with the highest priority were: **7, 2, 0, 6, and 8**.

Compared to the ranking obtained using the ICE framework, a change in order can be observed — hypotheses **7 and 2 moved up in the ranking**.  
This occurred because these hypotheses have a **higher Reach**, with values of **10 and 8**, respectively.

Therefore, the **RICE framework prioritizes ideas that affect a larger number of users**, even if their individual impact is slightly lower.

## 6. A/B Test Analysis

### 6.1 Cumulative Revenue

```python
orders.head()
```

```python
print('Experiment period:', orders['date'].min(), 'to', orders['date'].max())
```

```python
orders['date'] = pd.to_datetime(orders['date'])
```

```python
visits['date'] = pd.to_datetime(visits['date'])
```

```python
# creating a dataset with unique combinations of date and test group
datesgroup = orders[['date','group']].drop_duplicates()
```

```python
# calculating cumulative daily metrics for orders by test group

orderAggregated = datesgroup.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'],
            orders['group'] == x['group']
        )
    ].agg({
        'date' : 'max',
        'group': 'max',
        'transactionId' : pd.Series.nunique,
        'visitorId' : pd.Series.nunique,
        'revenue' : 'sum'
    }), 
    axis=1
).sort_values(by=['date','group'])
```

```python
visits.head()
```

```python
# calculating cumulative daily visitor metrics by test group

visitsAggregated = datesgroup.apply(
    lambda x: visits[
        np.logical_and(
            visits['date'] <= x['date'],
            visits['group'] == x['group']
        )
    ].agg({
        'date' : 'max',
        'group': 'max',
        'visits': 'sum'
    }), 
    axis=1
).sort_values(by=['date','group'])
```

```python
# merging the two datasets and assigning descriptive column names
cumulativedata = orderAggregated.merge(visitsAggregated, left_on= ['date','group'], right_on= ['date','group'])
cumulativedata.columns = ['date','group','orders','buyers','revenue','visitors']
cumulativedata.head()
```

```python
culumative_revenueA = cumulativedata[cumulativedata['group']== 'A'][['date','revenue','orders']]
culumative_revenueB = cumulativedata[cumulativedata['group']== 'B'][['date','revenue','orders']]
```

```python
plt.plot(culumative_revenueA['date'], culumative_revenueA['revenue'], label='Group A')
plt.plot(culumative_revenueB['date'], culumative_revenueB['revenue'], label='Group B')

plt.title('Cumulative Revenue by Group')
plt.xlabel('Date')
plt.ylabel('Cumulative Revenue')
plt.xticks(rotation=45)


plt.legend()
plt.grid(True)
plt.show()
```

### Cumulative Revenue Analysis

According to the chart, **Group B shows a faster growth in cumulative revenue** compared to Group A throughout the experiment period.

It is also possible to observe that **Group B presents revenue spikes at certain points in time**.  
This behavior may indicate an increase in the number of orders or the occurrence of **very high-value purchases**, which temporarily raise the total revenue.

These spikes suggest the need for a **more detailed investigation of potential outliers**, such as users who placed an unusually high number of orders or purchases with exceptionally high values.

### 6.2 Average Order Value

```python
plt.plot(culumative_revenueA['date'], culumative_revenueA['revenue']/culumative_revenueA['orders'], label='Group A')
plt.plot(culumative_revenueB['date'], culumative_revenueB['revenue']/culumative_revenueB['orders'], label='Group B')

plt.title('Cumulative Average Order Value by Group')
plt.xlabel('Date')
plt.ylabel('Cumulative Average Order Value')

plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```

### Average Order Value Analysis

During the first half of the test, the **average order value remained relatively stable for both groups**, without significant variations.

In the second half of the experiment, **Group B showed a noticeable spike in average order value**, suggesting a possible temporary increase in purchase value.

However, these occasional fluctuations may have been caused by **exceptionally high-value orders**, which can inflate the overall average.

Therefore, it is recommended to conduct an **outlier analysis** to determine whether the observed increase reflects a genuine change in user behavior or is simply the result of extreme values.

```python
merged_cumulativeRevenue = culumative_revenueA.merge(culumative_revenueB, left_on= 'date', right_on= 'date', how='left', suffixes=['A', 'B'])
merged_cumulativeRevenue.head()
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    merged_cumulativeRevenue['date'], 
    (merged_cumulativeRevenue['revenueB']/merged_cumulativeRevenue['ordersB']) /
    (merged_cumulativeRevenue['revenueA']/merged_cumulativeRevenue['ordersA']) - 1
)

plt.axhline(y=0, color='gray', linestyle='--')

plt.title('Relative Difference in Cumulative Average Order Value (B vs A)')
plt.xlabel('Date')
plt.ylabel('Relative Difference')

plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

### Relative Difference in Average Order Value

At several points during the experiment, the relative difference in cumulative average order value between groups A and B shows sharp spikes.

These fluctuations suggest the presence of high-value orders and potential outliers that directly affect the average order value.

This behavior reinforces the importance of conducting a detailed analysis of extreme orders in order to determine whether the observed increase represents a consistent trend or is simply the result of isolated high-value purchases.

### 6.3 Conversion Rate

```python
# calculating the cumulative conversion rate
cumulativedata['conversion'] = cumulativedata['orders'] / cumulativedata['visitors']
cumulativedata.head()
```

```python
CumulativeDataA = cumulativedata[cumulativedata['group']== 'A']
```

```python
CumulativeDataB = cumulativedata[cumulativedata['group']== 'B']
```

```python
plt.figure(figsize=(10,6))

plt.plot(CumulativeDataA['date'], CumulativeDataA['conversion'] * 100, label='Group A')
plt.plot(CumulativeDataB['date'], CumulativeDataB['conversion'] * 100, label='Group B')

plt.title('Conversion Rate by Group')
plt.xlabel('Date')
plt.ylabel('Conversion Rate (%)')

plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```

### Daily Conversion Rate Analysis

Overall, the conversion rates for groups A and B follow similar patterns throughout most of the experiment period, showing only small daily fluctuations. On some days, **Group B shows a slightly higher conversion rate**, while on others **Group A takes the lead**, without a clear and sustained trend.

These variations may be related to **daily traffic volume**, specific user behaviors on certain dates, or the impact of **external campaigns** that affect both groups equally. The absence of a consistent difference suggests that, based solely on daily conversion rates, it is not possible to conclude that one group significantly outperforms the other.

These observations highlight the need to **conduct statistical tests** to determine whether the observed differences reflect real effects or are simply the result of random variation in user behavior.

```python
mergedCumulativeConversions= CumulativeDataA[['date','conversion']].merge(CumulativeDataB[['date','conversion']], left_on='date', right_on='date',how='left', suffixes=['A','B'])
mergedCumulativeConversions.head()
```

```python
plt.plot(
    mergedCumulativeConversions['date'],
    mergedCumulativeConversions['conversionB'] / mergedCumulativeConversions['conversionA'] - 1,
    label='Relative Difference (B vs A)'
)

plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.axhline(y=-0.1, color='grey', linestyle='--', linewidth=1)

plt.title('Relative Difference in Conversion Rate (Group B vs Group A)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Relative Difference', fontsize=12)

plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

### Relative Difference in Daily Conversion Rate Analysis

During the first five days of the experiment, **Group B showed a lower conversion rate than Group A**, remaining below the reference line (0). This initial behavior may indicate a period of user adaptation to the tested variation or simply a natural fluctuation at the beginning of the experiment, when the data volume is still relatively small.

After this initial period, **the conversion rate of Group B increased and remained consistently above that of Group A**, fluctuating mainly between **0.10 and 0.20**. This means that, for most of the experiment, **Group B converted between 10% and 20% better than Group A**.

Although these values suggest a potential advantage for Group B, it is still necessary to verify whether this difference is **statistically significant**, since daily fluctuations may occur due to random variation in user behavior.

### 6.5 Statistical Significance Tests

```python
orders.head()
```

```python
visits.head()
```

```python
ordersByUsersA = orders[orders['group']== 'A'].groupby('visitorId', as_index=False).agg({'transactionId': pd.Series.nunique})
ordersByUsersA.columns =['userId','orders']
ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId': pd.Series.nunique})
ordersByUsersB.columns= ['userId','orders']
ordersByUsersA.head()
```

```python
sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='A']['visits'].sum() - len(ordersByUsersA['orders'])), name='orders')], axis=0)
sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='B']['visits'].sum() - len(ordersByUsersB['orders'])), name='orders')], axis=0)
```

```python
from scipy import stats as st

print('p-value', "{0:.3f}".format(st.mannwhitneyu(sampleA, sampleB)[1]))
print('Relative difference:', "{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

### Statistical Significance of the Difference in Conversion Rate (Raw Data)

The Mann–Whitney U test was applied to compare the conversion rates of groups A and B using the raw data. The test produced a p-value of 0.011, which is lower than the 5% significance level. Therefore, we reject the null hypothesis and conclude that there is a statistically significant difference between the conversion rates of the two groups.

The relative difference between the groups is +16%, indicating that Group B has a higher conversion rate than Group A.

This result suggests that the tested variation has a meaningful impact on user purchasing behavior.

```python
from scipy import stats

revenue_A = orders[orders['group'] == 'A']['revenue']
revenue_B = orders[orders['group'] == 'B']['revenue']

p_value = stats.mannwhitneyu(revenue_A, revenue_B).pvalue
relative_diff = revenue_B.mean() / revenue_A.mean() - 1

print(f'p-value: {p_value:.3f}')
print(f'Relative difference: {relative_diff:.3f}')
```

### Statistical Significance of the Difference in Average Order Value (Raw Data)

The Mann–Whitney U test applied to the order values of groups A and B produced a **p-value of 0.862**, which is much higher than the **5% significance level**. Therefore, there is **no statistical evidence** to conclude that the average order value differs between the two groups.

Although the **raw average order value of Group B appears to be 27.8% higher than that of Group A**, this difference is **not statistically significant**. This pattern suggests the influence of **outliers**, which were previously identified in the scatter plots and in the order value percentiles.

Thus, the observed difference in average order value is most likely caused by **a small number of unusually large orders**, rather than a real effect of the tested change.

This indicates that the tested variation does not significantly affect the value of individual purchases.

```python
orders.head()
```

```python
# filtering users with more than 2 orders
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > 2]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > 2]['userId']
    ],
    axis=0
)
```

```python
# identifying users with orders above 830 (potential revenue outliers)
usersWithExpensiveOrders = orders[orders['revenue'] > 830]['visitorId']
```

```python
abnormalUsers = pd.concat([usersWithManyOrders,usersWithExpensiveOrders], axis=0).drop_duplicates().sort_values()
```

```python
print(abnormalUsers.head(5))
print(abnormalUsers.shape)
```

```python
# preparing filtered samples for conversion rate comparison between groups


sampleAFiltered = pd.concat(
    [
        ordersByUsersA[np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visits[visits['group']=='A']['visits'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders'
        )
    ],
    axis=0
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visits[visits['group']=='B']['visits'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders'
        )
    ],
    axis=0
)
```

```python
print("p-value", "{0:.3f}".format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print("Relative difference", "{0:.3f}".format(sampleBFiltered.mean() / sampleAFiltered.mean() - 1))
```

### Statistical Significance of the Difference in Conversion Rate (Filtered Data)

After removing anomalous users — those with more than 2 orders and those who placed orders with a value above 830.3 — the Mann–Whitney U test was recalculated to evaluate the conversion rates between groups A and B. The resulting p-value was **0.006**, which is below the **5% significance level**, indicating that the difference between the groups remains **statistically significant even after filtering the data**.

The relative difference between the group averages is **+19.2%**, showing that **Group B consistently achieves a higher conversion rate than Group A**. The removal of outliers increased the reliability of the analysis and confirmed that the advantage observed for Group B was **not caused by extreme values**.  
                                                                                                                                                        This result strengthens the evidence that the tested variation positively affects user conversion behavior.

```python
print('p-value','{0:.3f}'.format(stats.mannwhitneyu(
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'],
    orders[np.logical_and(
        orders['group']=='B',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'])[1]))

print('Relative difference','{0:.3f}'.format(
    orders[np.logical_and(
        orders['group']=='B',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean() /
    orders[np.logical_and(
        orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean() - 1))
```

### Statistical Significance of the Difference in Average Order Value (Filtered Data)

After removing users with anomalous behavior — those with more than 2 orders or with orders above 830.3 — the Mann–Whitney U test was applied to compare the average order value between groups A and B. The resulting p-value was **0.877**, which is much higher than the **5% significance level**. This indicates that there is **no statistically significant difference** in average order value between the two groups.

The relative difference between the averages also supports this conclusion: **Group B shows an average order value 1.4% lower than Group A**, a variation that is negligible from both a statistical and business perspective. Therefore, the experiment **did not affect the average order value**, but only the **conversion rate**.

This suggests that the tested variation increases the likelihood of purchase without changing the amount users spend per order.
