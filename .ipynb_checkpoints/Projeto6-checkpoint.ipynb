## 3. Data Description

This analysis uses three datasets containing information about business hypotheses, user visits, and customer orders during the A/B test.

### hypotheses_us.csv

This dataset contains the list of marketing and product hypotheses designed to increase the revenue of the online store. Each hypothesis was evaluated using four criteria that support prioritization frameworks.

Variables:

- **Hypothesis** — brief description of the proposed improvement or experiment.
- **Reach** — estimated number of users affected by the change (scale from 1 to 10).
- **Impact** — expected impact on user behavior if the hypothesis is implemented (scale from 1 to 10).
- **Confidence** — level of confidence in the hypothesis (scale from 1 to 10).
- **Effort** — estimated resources required to implement and test the hypothesis (scale from 1 to 10).

---

### orders_us.csv

This dataset contains information about orders placed during the A/B test.

Variables:

- **transactionId** — unique identifier of each order.
- **visitorId** — unique identifier of the user who placed the order.
- **date** — date when the order was made.
- **revenue** — revenue generated by the order.
- **group** — A/B test group to which the user belongs (A or B).

---

### visits_us.csv

This dataset contains the number of daily visits for each test group.

Variables:

- **date** — date of the recorded visits.
- **group** — A/B test group (A or B).
- **visits** — total number of visits recorded on that date for the corresponding group.

---

Before performing the analysis, the datasets will be inspected and preprocessed to ensure data quality, including checking for missing values, incorrect data types, duplicates, and potential inconsistencies such as users appearing in both experimental groups.
